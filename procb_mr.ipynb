```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# columns: 1=p, 2=H2_mr, 3=CH4_mr, 4=CO_mr
procb_bio_data = pd.read_csv('/data/jwm212/PALEOjulia/PALEOexamples/src/mphys2022/atmchem_methan_CO_JamesM/H2CH4COmr_bio_procb.csv', sep='\t', header=None, skiprows=1)
procb_abiotic_data = pd.read_csv('/data/jwm212/PALEOjulia/PALEOexamples/src/mphys2022/atmchem_methan_CO_JamesM/H2CH4COmr_abiotic_procb.csv', sep='\t', header=None, skiprows=1)
```

```python
def Extract(output):
    data = [np.array(output[1]),
            np.array(output[2]),
            np.array(output[3]),
            np.array(output[4])]
    for i in data:
        [float(j) for j in i]
        
    for i in data:    
        i.tolist().reverse()
    
    p = data[0]
    h2_mr = data[1]
    ch4_mr = data[2]
    co_mr = data[3]
    
    z = []
    for i in p:
        z.append(((5e3)*math.log(1e5/i))/1e3)
        
    return z, h2_mr, ch4_mr, co_mr
    
```

```python
proc_bio_z, proc_bio_H2, proc_bio_CH4, proc_bio_CO = Extract(procb_bio_data)[0], Extract(procb_bio_data)[1], Extract(procb_bio_data)[2], Extract(procb_bio_data)[3]
proc_abiotic_z, proc_abiotic_H2, proc_abiotic_CH4, proc_abiotic_CO = Extract(procb_abiotic_data)[0], Extract(procb_abiotic_data)[1], Extract(procb_abiotic_data)[2], Extract(procb_abiotic_data)[3]
```

```python
fig = plt.figure(figsize=(10, 6), dpi= 300, facecolor='w', edgecolor='k')
ax = fig.add_subplot(111)
ax.plot(proc_bio_H2,proc_bio_z,color='blue', label=r'$H_2$',linewidth=3)
ax.plot(proc_abiotic_H2,proc_abiotic_z,color='blue',linestyle='dashed',linewidth=3)

ax.plot(proc_bio_CH4,proc_bio_z,color='green', label=r'$CH_4$',linewidth=3)
ax.plot(proc_abiotic_CH4,proc_abiotic_z,color='green',linestyle='dashed',linewidth=3)

ax.plot(proc_bio_CO,proc_bio_z,color='orange', label=r'$CO$',linewidth=3)
ax.plot(proc_abiotic_CO,proc_abiotic_z,color='orange',linestyle='dashed',linewidth=3)

ax.plot(0,0,color='black',label='Biotic',linewidth=3)
ax.plot(0,0,color='black',label="Abiotic",linestyle='dashed',linewidth=3)

ax.set_axisbelow(True)
for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(2)

for tick in ax.get_xticklabels():
    tick.set_fontname("Serif")
    tick.set_fontsize(24)
for tick in ax.get_yticklabels():
    tick.set_fontname("Serif")
    tick.set_fontsize(24)
    
ax.xaxis.set_tick_params(width=2)
ax.yaxis.set_tick_params(width=2)

ax.set_xscale('log')
ax.set_xlim(xmin=1e-7,xmax=1e-2)
ax.set_ylim(ymin=0,ymax=80)
ax.yaxis.set_ticks([0,20,40,60,80])

plt.xlabel(r'Volume Mixing Ratio ($m^3$/$m^3$)', fontname='Serif', fontsize=24)
plt.ylabel('Height ($km$)', fontname='serif', fontsize=24)

ax.legend(fontsize=19,loc='lower left')
#plt.savefig('procb_mr.jpg',bbox_inches='tight')
ax.plot()
```
