```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import gridspec
from matplotlib.ticker import MaxNLocator
from scipy.optimize import curve_fit
#import astropy.constants as const
import csv
import math
import numpy
```

```python
bio = np.array(pd.read_csv('psg_rad_biotic.txt',
                             sep='  ', header=21, skiprows=0))
abio = np.array(pd.read_csv('psg_rad_abiotic.txt',
                             sep='  ', header=21, skiprows=0))
```

```python
bio_wave, bio_noise, bio_transit = bio[:,0], bio[:,2], bio[:,4]
abio_wave, bio_noise, abio_transit = abio[:,0], abio[:,2], abio[:,4]
```

```python
xch4 = [0.9, 1.0, 1.1, 1.2, 1.6, 2.5, 3.5, 7.5]
ych4 = [3.585e-3, 3.585e-3, 3.588e-3, 3.588e-3, 3.588e-3, 3.59e-3, 3.6e-3, 3.6e-3]
```

```python
fig = plt.figure(figsize = (9,9), dpi= 500, facecolor='w', edgecolor='k')
# set height of each subplot as 8
fig.set_figheight(3)
 
# set width of each subplot as 8
fig.set_figwidth(7)

# create grid for different subplots
grid = gridspec.GridSpec(ncols=1, nrows=2,
                         hspace=0.3, height_ratios=[4, 1])

ax0 = fig.add_subplot(grid[0])
ax1 = fig.add_subplot(grid[1],sharex=ax0)
plt.setp(ax0.get_xticklabels(), visible=False)

ax0.plot(bio_wave,bio_transit,label = "Biotic", color='red',alpha=0.7,linewidth=1)
ax0.plot(abio_wave,abio_transit, label = "Abiotic", color='skyblue',alpha=1,linewidth=0.5)
ax1.plot(bio_wave,bio_transit-abio_transit, label = "(Biotic - Abiotic)",color='black',alpha=0.5,linewidth=1)
#ax0.scatter(xch4, ych4, marker="*", color="green", label=r"$CH_4$",s=200)
#ax0.scatter(0,0, marker="D", color="orange", label=r"$CO$", s=150)
#ax0.scatter(0,0, marker="^", color="purple", label=r"$CO_2$", s=150)

ax0.set_axisbelow(True)
for axis in ['top','bottom','left','right']:
  ax0.spines[axis].set_linewidth(1)
ax1.set_axisbelow(True)
for axis in ['top','bottom','left','right']:
  ax1.spines[axis].set_linewidth(1)

for tick in ax0.get_xticklabels():
    tick.set_fontname("Serif")
    tick.set_fontsize(10)
for tick in ax0.get_yticklabels():
    tick.set_fontname("Serif")
    tick.set_fontsize(10)
    
for tick in ax1.get_xticklabels():
    tick.set_fontname("Serif")
    tick.set_fontsize(10)
for tick in ax1.get_yticklabels():
    tick.set_fontname("Serif")
    tick.set_fontsize(10)

ax0.xaxis.set_tick_params(width=1)
ax0.yaxis.set_tick_params(width=1)
ax1.xaxis.set_tick_params(width=1)
ax1.yaxis.set_tick_params(width=1)


ax0.set_ylim(0.00357,0.003615)
#ax1.yaxis.set_ticks([0.0,1.0e-5])  
ax0.set_xlim(1,5.3)

#ax0.plot(line1CO2x,line1CO2y,color='black')
#ax0.text(1.9, 0.003572, '$CO_2$', fontsize = 20)
#ax0.set_yscale("log")
ax0.set_xscale("log")
ax1.set_xscale("log")
ax0.yaxis.set_ticks([3.58e-3, 3.60e-3]) 
ax1.xaxis.set_ticks([1,2,3,4,5]) 
ax1.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
ax0.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
ax1.set_xlabel(r"Wavelength, $\mu$m", fontname='serif', fontsize=10)
ax0.set_ylabel(r'$(R_p/R_s)^2$', fontname='serif', fontsize=10, labelpad=0)
ax1.set_ylabel("Residual", fontname='serif', fontsize=10, rotation=0, labelpad=30)
ax1.yaxis.offsetText.set_fontsize(10)
ax0.legend(fontsize=10,loc='upper left')
ax1.legend(fontsize=8,loc='upper right')
plt.savefig('transmission_rel_final.jpg',bbox_inches='tight')
plt.show()
```
