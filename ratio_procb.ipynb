```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
h2 = [1.0e11, 3.0e11, 9.0e11, 2.0e12, 5.0e12, 8.0e12, 1.0e13, 4.0e13, 7.0e13, 1.0e14]
p_bio = [0.2891360223470617, 0.2881647266800431, 0.5091647173808463, 0.34439275700290795, 0.20742860628071041, 0.17461193242615347, 0.16447872679482742, 0.12404821822655955, 0.1110234793168371, 0.10745174180878632]
p_abio = [1283.1218144412508, 867.5523320796959, 268.75109280490847, 284.9536056100937, 62.04256058403794, 51.889586851170755, 48.61340953157152, 354.1002609139183, 387.6135264269688, 405.60890773425706] #5th value suspicious
a_bio = pd.read_csv('../../DATA/ratio_bio_archean_Earth.csv', sep=',', header=None)
a_abio = pd.read_csv('../../DATA/ratio_abiotic_archean_Earth.csv', sep=',', header=None)

```

```python
fig = plt.figure(figsize=(5, 4), dpi= 500, facecolor='w', edgecolor='k')
ax = fig.add_subplot(111)
ax.scatter(h2, a_abio, color='skyblue',alpha=0.3)
ax.plot(h2, a_abio, color='skyblue', linewidth=2,alpha=0.3)
ax.plot(h2, p_abio, color='skyblue', label='Abiotic', linewidth=2)
ax.scatter(h2, p_abio, color='skyblue')

ax.scatter(h2, a_bio, color='red',alpha=0.3)
ax.plot(h2, a_bio, color='red', linewidth=2,alpha=0.3)
ax.plot(h2, p_bio, color='red', label='Biotic', linewidth=2)
ax.scatter(h2, p_bio, color='red')

ax.plot(0,0,color='gray',label='Archean Earth',alpha=0.3)



ax.set_axisbelow(True)
for axis in ['top','bottom','left','right']:
  ax.spines[axis].set_linewidth(1)

for tick in ax.get_xticklabels():
    tick.set_fontname("Serif")
    tick.set_fontsize(11)
for tick in ax.get_yticklabels():
    tick.set_fontname("Serif")
    tick.set_fontsize(11)
    
ax.xaxis.set_tick_params(width=1)
ax.yaxis.set_tick_params(width=1)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim(ymin=1e-2, ymax=1e4)
ax.set_xlim(xmin=1e11,xmax=1e14)
ax.set_yscale('log')
plt.xlabel(r'Outgassed $H_2$ flux (mol yr$^{-1}$)', fontname='Serif', fontsize=11)
plt.ylabel('CO/CH4 ratio', fontname='serif', fontsize=11)

plt.legend(loc=(0.46,0.74),fontsize=11)
plt.savefig('procb_ratio.jpg',bbox_inches='tight')
plt.show()
```
